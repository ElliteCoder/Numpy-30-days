# **LINEAR ALGEBRA**

```python
import numpy as np

# ============================================================
# LINEAR ALGEBRA IN NUMPY
# ============================================================
# Linear Algebra is the mathematics of vectors and matrices.
# It is the backbone of Machine Learning and Deep Learning.
# Every single ML operation at its core is linear algebra.
#
# WHY IT IS IMPORTANT IN ML:
# Neural Network forward pass  → X @ W + b
# PCA                          → eigenvalues and eigenvectors
# Solving optimization         → np.linalg.solve
# Measuring similarity         → dot product
# Image transformations        → matrix multiplication
#
# WITHOUT LINEAR ALGEBRA THERE IS NO ML. PERIOD.
#
# NumPy makes linear algebra:
# → Fast    (runs in C under the hood)
# → Simple  (one line instead of 50 lines)
# → Accurate (no manual calculation errors)
# ============================================================

# ============================================================
# MATRIX MULTIPLICATION RULES (Keith Galli)
# ============================================================
# RULE: columns of FIRST matrix must match rows of SECOND matrix
#
# A.shape = (2,3) → 2 rows, 3 columns
# B.shape = (3,2) → 3 rows, 2 columns
#
# columns of A = 3 ✅ matches rows of B = 3
# if they dont match → ERROR
#
# Result shape = (rows of A, columns of B)
# Result shape = (2,2)
# ============================================================

a = np.ones((2,3))
print(a)
# [[1. 1. 1.]
#  [1. 1. 1.]]

b = np.full((3,2),2)
print(b)
# [[2 2]
#  [2 2]
#  [2 2]]

# Matrix Multiplication
# Formula = np.matmul(first array, second array)
# OR use @ operator → a @ b (exact same thing, cleaner)
result = np.matmul(a,b)
print(result)
# [[6. 6.]
#  [6. 6.]]

# WHY 6?
# each element = sum of (row of A multiplied by column of B)
# row of A    = [1, 1, 1]
# column of B = [2, 2, 2]
# (1x2) + (1x2) + (1x2) = 2 + 2 + 2 = 6
#
# This exact operation X @ W is what every
# dense layer in a neural network does
# X = input data
# W = weights
# result = predictions
```

```python
a = np.ones((2,3))
print(a)

b = np.full((3,2 ),2)
print (b)
# To Multiply the two array in linear algebra 
# We use matmul (Matrix Multiplication) to multiply the matrix
np.matmul(a,b)
```
